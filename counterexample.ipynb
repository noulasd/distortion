```python
factor(E0.cardinality())

```

Define the curve over F_p^6 so it contains all 4-torsion

```python
k = 6
E= EllipticCurve(GF(p^k),[-a,b])
```

```python
E
```

```python
factor(E.cardinality())
```

```python
torsion_4 = E(0).division_points(4)
torsion_4
```

```python
P = torsion_4[1]
P
```

```python
psi = E.isogeny(P, algorithm='factored')
psi
```

```python
E.j_invariant()
```

```python
E_image = psi.codomain()
E_image.j_invariant()
```

```python
gamma = E_image.isomorphism_to(E)
gamma
```

Coordinates of the isomorphism gamma:

```python
gamma.rational_maps()
```

Coordinates of the isogeny psi:

```python
psi.rational_maps()
```

```python
torsion_2 = E(0).division_points(2)
torsion_2
```

Loop over the 2-torsion points by the isogeny composed with the isomorphism phi = gamma \circ psi

```python
for P in torsion_2:
    print( gamma(psi(P)))
```
