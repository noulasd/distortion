```python
########################## WE CONSTRUCT THE DISTORTION MAP BY VELU AS E->E/(P) FOR A TORSION 9 POINT P, THIS QUOTIENT ACTUALLY COINCIDES WITH E #####################################

P = E(0).division_points(9)[1]


phi = E.isogeny(P,algorithm='factored')
phi
```

Below we showcase why the conductor of our Elliptic curve E is 3

```python
# 1. Get the j-invariant
j_val = E.j_invariant()

# 2. Get the Class Polynomial for D = -27
# This polynomial describes ALL curves with End(E) of discriminant -27
H = hilbert_class_polynomial(-27)

# 3. Check if your j-invariant is a root
R = PolynomialRing(GF(q), 'x')
is_correct_conductor = R(H)(j_val) == 0

print(f"j-invariant: {j_val}")
print(f"Does it match conductor 3 (D=-27)? {is_correct_conductor}")
```
